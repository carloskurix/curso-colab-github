```python
# Ejemplo de curso "Hola Mundo" en Colab/GitHub
# Tema random: Cálculo del área de un círculo

import math

def calcular_area_circulo(radio):
    """
    Calcula el área de un círculo dado su radio.
    """
    area = math.pi * (radio ** 2)
    return area

# Definir un radio de ejemplo
radio_ejemplo = 5

# Calcular el área
area_resultante = calcular_area_circulo(radio_ejemplo)

# Imprimir el resultado
print(f"El área del círculo con radio {radio_ejemplo} es: {area_resultante}")
```
